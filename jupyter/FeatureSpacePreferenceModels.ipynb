```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ood_df = pd.read_csv('/users/lucelo/UQLRM/uqlrm/scripts/slurm/metadata_llama_rw_infer_v0_.tsv', sep='\t', header=0)
```

```python
ood_df.head()
```

```python
ood_df['RewardScore'].hist(bins=100)
```

```python
ood_df[ood_df['Preference'] == 'chosen']['RewardScore'].hist(bins=100, alpha=0.5, label='chosen')
ood_df[ood_df['Preference'] == 'rejected']['RewardScore'].hist(bins=100, alpha=0.5, label='rejected')
plt.legend()
plt.title("Chosen/Rejected - Llama2")
```

```python
ood_df[ood_df['Dataset'] == 'train']['RewardScore'].hist(bins=100, alpha=0.5, label='train')
ood_df[ood_df['Dataset'] == 'ood']['RewardScore'].hist(bins=100, alpha=0.5, label='ood')
ood_df[ood_df['Dataset'] == 'shuffled']['RewardScore'].hist(bins=100, alpha=0.5, label='shuffled')
plt.legend()
plt.title("Reward Scores - Llama7B")
```
